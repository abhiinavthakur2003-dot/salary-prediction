```python
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\thaku\Desktop\sk-learn\Salary_dataset_extended_2.csv")
```

```python
x=data[['YearsExperience']]
```

```python
x.head()
```

```python
x.shape
```

```python
y=data['Salary']
```

```python
y.head()
```

```python
y.shape
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict([[3]])
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred=model.predict(x_test)
```

```python
r2_score(y_test,y_pred)*100
```

```python
r2_score(y_train,model.predict(x_train))*100
```

```python
import matplotlib.pyplot as plt 
plt.scatter(x_train,y_train,color='skyblue',label='Training Data')
plt.scatter(x_test,y_test,color='orange',label='Testing Data')
plt.plot(x_test,y_pred,color='red',label='BFT')
plt.legend()
plt.show()
```
